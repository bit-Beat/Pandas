```python
import numpy as np
import pandas as pd

##머신러닝 라이브러리이므로 판다스 실습 시 굳이 필요 X
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier
```

```python
df= pd.read_csv('us_new_born_2million.csv') #CSV 엑셀파일 불러와서 df변수에 대입
```

```python
print(df.columns) # df의 컬럼정보 출력
print("\n============================================================================================================\n")
print(df.index) # df의 인덱스 정보 출력
print("\n============================================================================================================\n")
print(df.info()) # df의 정보 출력
print("\n============================================================================================================\n")
print(df.isna()) #df의 null정보 출력
print("\n============================================================================================================\n")
print(df.isna().sum()) #df의 null 개수 출력
print("\n============================================================================================================\n")
print(df.isna().sum().sort_values(ascending=False)) #df의 null개수를 내림차순 정렬해서 출력
print("\n============================================================================================================\n")
print(df.index.size) # df의 인덱스 크기 출력
print("\n============================================================================================================\n")
print(df.isna().sum() / df.index.size) # null 개수의 행의 길이(인덱스 사이즈)를 나누어주면 null의 비율이 나온다. 
print("\n============================================================================================================\n")
print(df.isna().sum() / df.index.size>0.7) # null의 비율이 0.7(70%) 이상인 열들을 출력
print("\n============================================================================================================\n")
print("null 개수의 평균값 : ",(df.isna().sum() / df.index.size).mean())
print("null 개수의 표준편차 : ",(df.isna().sum() / df.index.size).std())
print("\n============================================================================================================\n")
print(df.isna().sum()[df.isna().sum() > (df.isna().sum() / df.index.size).mean()].index) #null의 개수가 null 개수의 평균값보다 높은 열 출력
print("\n============================================================================================================\n")
print(df[['weight_pounds','state']]) #weight 와 state 열 출력
print(df.drop(['weight_pounds','state'],axis=1)) #weight와 state 행 삭제 inplace=True를 매개변수로 줄 시 df의 결과 즉각 반영
print("\n============================================================================================================\n")
print(df.isna().sum(axis=1)) #axis = 1을 주어 행을 기준(index기준)으로 null의 개수 합계를 출력
print(np.sum(df.isna().sum(axis=1))) #위 결과 값을 모두 더한다.
print("\n============================================================================================================\n")
print(df.isna().sum(axis=1)[(df.isna().sum(axis=1) / df.columns.size) > 0.3].index) # null의 개수가 로우를 기준으로 30%비율이 넘는 row(index)를 출력
print("\n============================================================================================================\n")
print(df.shape)
print(df.drop(df.isna().sum(axis=1)[(df.isna().sum(axis=1) / df.columns.size) > 0.3].index,axis=0).shape)
#shape를 확인해보면 전 과 후의 shape가 변하였다. 즉, 30%비율니 넘는 row(index)가 삭제된 걸 확인 할 수 있다.
print("\n============================================================================================================\n")
df1 = df.copy() #df1의 df복사
print("<Before>df1.weighte_pounds.isna().sum():",df1.weight_pounds.isna().sum()) #weight_pounds의 null 합계
#df1.weight_pounds.fillna(10) #df1의 weigh_pounds 의 10으로 값을 채운다.
df1.weight_pounds.fillna(df1.weight_pounds.mean(), inplace=True) #weight_pounds의 평균값으로 채운다.
print("<After>df1.weight_pounds.isna().sum()",df1.weight_pounds.isna().sum()) #평균으로 채운 후 결과 출력
print("\n============================================================================================================\n")
df1 = df.copy()
print(df1.loc[df1.is_male,'weight_pounds'].mean()) #남자의 몸무게 평균. 
print(df1.loc[~df1.is_male,'weight_pounds'].mean()) #여자의 몸무게 평균 ~은 False -> True로 전환 True -> False로 전환
print("\n============================================================================================================\n")
x = pd.Series(np.arange(10)) # 0 ~9 값의 시리즈 생성
df1.weight_pounds[df1.is_male & df1.weight_pounds.isna()] = df1.loc[df1.is_male,'weight_pounds'].mean() #남자중의 weight_pounds의 칸이 비어있다면 남자의 몸무게 평균으로 값을 채워넣는다.
df1.weight_pounds[~df1.is_male & df1.weight_pounds.isna()] = df1.loc[~df1.is_male,'weight_pounds'].mean() #여자중의 weight_pounds의 칸이 비어있다면 여자의 몸무게 평균으로 값을 채워넣는다
print( df1.weight_pounds.isna().sum() ) #값이 모두 채워져서 0으로 출력
print("\n============================================================================================================\n")
df1 = df.copy()
print(np.where(df1.weight_pounds.isna() , df1.loc[df1.is_male,'weight_pounds'].mean() , df1.weight_pounds)) #np.where을 이용해서 남자중의 weight_pounds의 칸이 비어있다면 남자의 몸무게 평균으로 값을 채워넣는다.
print("\n============================================================================================================\n")
data = df1.loc[~df1.cigarette_use.isna() | ~df1.alcohol_use.isna() , :] #cigarette 와 alcohol 이 비어 있지 않은 컬럼
print(data.isna().sum()) #cigarette, alcohol이 0
print("\n============================================================================================================\n")
print(data.dtypes) # object는 타입을 바꿔줘야 함
print("\n============================================================================================================\n")
data1 = data[['plurality','mother_race','mother_age','father_age','father_race','cigarette_use']]
print(data1)
print("<Before>:",data1.dtypes) #타입 출력
data1.cigarette_use = data1.cigarette_use.astype('int64') #object에서 int64로 타입 변경
print("<After>:",data1.dtypes) #타입 출력
print("\n============================================================================================================\n")
```
