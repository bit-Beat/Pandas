```python
import numpy as np
import pandas as pd

import plotly.express as px

import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output, State
```

```python
df = px.data.gapminder()
```

```python
app = dash.Dash(__name__,external_stylesheets=[dbc.themes.SLATE])
```

```python
radio = dcc.RadioItems( id="radio", options =[{'label':i,'value':i} for i in df.continent.unique()], style={'width':'380px'} )
drop = dcc.Dropdown( id="drop", style={'width':'400px'})
range_slider = dcc.RangeSlider( id="range_slider",min=df.year.min(),max=df.year.max(),marks = {int(i) :str(i)+'년도' for i in df.year.unique()})
bar_graph =dcc.Graph(id='bar_graph',style={'border':'3px dotted white','margin-right':'20px'} )
input_box = dcc.Input(id="input_box", type='test',placeholder='Country Input',style={'width':'90%'})
input_button = html.Button("👍",id="input_button",n_clicks=0,style={'width':'8%','margin-left':'2px'})

main=dbc.Card(
    [
        dbc.CardBody(
            bar_graph
        )
    ]
)

control = dbc.Card(
    [
        dbc.CardBody(
            [
                dbc.Card([dbc.CardBody([html.Label("대륙을 선택하시오."),radio])],style={'margin-bottom':'12px'}),
                dbc.Card([dbc.CardBody([html.Label("나라를 선택하시오."),drop])],style={'margin-bottom':'12px'}),
                dbc.Card([dbc.CardBody([html.Label("연도를 선택하시오."),range_slider])],style={'margin-bottom':'12px'}),
                dbc.Card([dbc.CardBody([input_box,input_button])]),
            ]
        )
    ]
)

main1 = dbc.CardDeck(
    [
        control,
        bar_graph
    ],style={'height':'100vh'}
)
```

```python
app.layout = html.Div(
    [
        main1
    ],style={'overflow':'hidden'}
)
```

```python
@app.callback(
    Output('drop','options'),
    Input('radio','value')
)
def update_drop_down(value):
    options = [{'label':i,'value':i} for i in df.loc[df.continent == value,"country"].unique()]
    return options

@app.callback(
    Output("bar_graph",'figure'),
    Input('drop','value'),
    Input('range_slider','value')
)
def update_bar_graph(country,year):
    title = "{}의 {}년 - {}년 인구변화".format(country,year[0],year[0])
    title = f'{country}의 {year[0]}년 ~ {year[1]}년 인구변화'
    data = df.loc[(df.country==country) & ((df.year >= year[0]) & (df.year <= year[1])),['year','pop']]
    fig = px.bar(data,x='year',y='pop',color='pop',title=title)
    return fig
```

```python
app.run_server()
```
